<div style="text-align: justify; font-size: 20px; color: purple; font-family: Garamond; font-weight: bold">
   EXPLORATORY DATA ANALYSIS
</div>
<div style="text-align: justify; font-size: 16px; color: #000000; font-family: Garamond;">
    We will explore the data to uncover patterns and trends in the dataset. Assess total and average performance metrics by agent. Examine conversion rates and dropped calls to identify areas of improvement.

```python
# Trend Analysis: Total Calls, Calls Reached, Deals Closed Over Time
df.groupby('Date')[['Total Calls', 'Calls Reached', 'Deals Closed']].sum().sort_index().plot(
    kind='line', 
    figsize=(10, 6)
)
plt.title('Trend Analysis Over Time')
plt.xlabel('Date')
plt.ylabel('Count')
plt.show()
print("\n")

# Performance Analysis by Agent (Sorted based on the highest deal values)
agent_performance = df.groupby('Name')[['Total Calls', 'Calls Reached', 'Deals Closed', 'Deal Value ($)']].sum()
agent_performance_sorted = agent_performance.sort_values(by='Deal Value ($)', ascending=False)
print(agent_performance_sorted)
print("\n")
print("\n")

# Conversion and Drop Rates by Agent (Sorted based on the highest call conversion rate (%))
conversion_drop_rates = df.groupby('Name')[['Call Conversion Rate (%)', 'Call Drop Rate (%)']].mean()
conversion_drop_rates_sorted = conversion_drop_rates.sort_values(by='Call Conversion Rate (%)', ascending=False)
print(conversion_drop_rates_sorted)
```

```python
# Bar Chart: Total Calls by Agent (Sorted by highest total calls)
sorted_total_calls = agent_performance['Total Calls'].sort_values(ascending=False)
sorted_total_calls.plot(kind='bar', figsize=(8, 5), color = 'purple', title='Total Calls by Agent')
plt.ylabel('Total Calls')
plt.xlabel('Agent')
plt.show()

# Top 7 Agents by Conversion Rate
top_7_agents_conversion_rate = conversion_drop_rates['Call Conversion Rate (%)'].sort_values(ascending=False).head(7)
print("Top 7 Agents by Call Conversion Rate:")
print(top_7_agents_conversion_rate)
top_7_agents_conversion_rate.plot(kind='bar', figsize=(8, 5), color = '#8B008B', title='Big 7 Agents by Call Conversion Rate')
plt.ylabel('Conversion Rate (%)')
plt.xlabel('Agent')
plt.show()

# Top 7 Agenrs by Deal Value
top_7_agents_deal_value = agent_performance['Deal Value ($)'].sort_values(ascending=False).head(7)
print("Top 7 Agents by Deal Value:")
top_7_agents_deal_value.plot(kind='bar', figsize=(8, 5), title='Big 7 Agents by Deal Value', color='violet')
plt.ylabel('Deal Value ($)')
plt.xlabel('Agent')
plt.show()

```

```python
#DEALS CLOSED BY CALL DURATION
bins = [0, 200, 400, 600, 800]  # Define bins 
labels = ['0-200 sec', '201-400 sec', '401-600 sec', '601-800 sec']

# Create a new column for duration categories
df['Duration Category'] = pd.cut(df['Average Duration (sec)'], bins=bins, labels=labels, include_lowest=True)

# Aggregate Deals Closed by Duration Category
duration_analysis = df.groupby('Duration Category', observed = False)['Deals Closed'].sum().reset_index()
print(duration_analysis)

# Bar Chart for Deals Closed by Call Duration Categories
plt.figure(figsize=(10, 6))
plt.bar(duration_analysis['Duration Category'], duration_analysis['Deals Closed'], color='#CEA2FD')
plt.title('Deals Closed by Call Duration Categories')
plt.xlabel('Call Duration Categories')
plt.ylabel('Deals Closed')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
